```python
#url = "https://www.airbnb.com/s/Boston--MA--United-States/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&query=Boston%2C%20MA%2C%20United%20States&place_id=ChIJGzE9DS1l44kRoOhiASS_fHg&source=structured_search_input_header&search_type=autocomplete_click"
url = "https://www.airbnb.com/s/Denver--CO--United-States/homes?refinementPaths%5B%5D=%2Fhomes&query=Denver%2C%20CO%2C%20United%20States&place_id=ChIJzxcfI6qAa4cR1jaKJ_j0jhE&refinement_paths%5B%5D=%2Fhomes&source=structured_search_input_header&search_type=autocomplete_click"
r = requests.get(url)
c = r.content
soup = bs(c, features = 'lxml')

# find a list using find_all
rooms = soup.findAll("div", {"class": "_8ssblpx"})

result = []
for room in rooms:
    result.append(room)
```
